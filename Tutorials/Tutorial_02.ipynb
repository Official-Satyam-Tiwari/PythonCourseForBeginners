````python
# There is something called Frozen Sets in python, to support more constraints in Sets data structure.
````

````python
# Creating a Set
people = {"Jay", "Idrish", "Archi"}

print("People:", end = " ")
print(people)

# This will add Daxit in the set
people.add("Daxit")

# Adding elements to the set using iterator
for i in range(1, 6):
    people.add(i)

print("\nSet after adding element:", end = " ")
print(people)
````

````python
people = {"Jay", "Idrish", "Archil"}
vampires = {"Karan", "Arjun"}
dracula = {"Deepanshu", "Raju"}

# Union using union() function
population = people.union(vampires)

print("Union using union() function")
print(population)

# Union using "|" operator
population = people|dracula
print("\nUnion using '|' operator")
print(population)
````

````python
set1 = set()
set2 = set()

for i in range(5):
    set1.add(i)

for i in range(3,9):
    set2.add(i)

# Intersection using intersection() function
set3 = set1.intersection(set2)

print("Intersection using intersection() function")
print(set3)

# Intersection using "&" operator
set3 = set1 & set2
print("\nIntersection using '&' operator")
print(set3)
````

````python
set1 = set()
set2 = set()

for i in range(5):
    set1.add(i)

for i in range(3,9):
    set2.add(i)

# Difference of two sets using difference() function
set3 = set1.difference(set2)
print(" Difference of two sets using difference() function")
print(set3)

# Difference of two sets using '-' operator
set3 = set1 - set2
print("\nDifference of two sets using '-' operator")
print(set3)
````

````python
set1 = {1,2,3,4,5,6}
print("Initial set")
print(set1)

# This method will remove all the elements of the set
set1.clear()
print("\nSet after using clear() function")
print(set1)
````

## Operators

````python
print("\n--- 1. ARITHMETIC OPERATORS ---")
a, b = 10, 3
print("a =", a, ", b =", b)
print("Addition (a + b):", a + b)
print("Subtraction (a - b):", a - b)
print("Multiplication (a * b):", a * b)
print("Division (a / b):", a / b)
print("Floor Division (a // b):", a // b)
print("Modulus (a % b):", a % b)
print("Exponent (a ** b):", a ** b)

print("\n--- 2. COMPARISON OPERATORS ---")
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)

print("\n--- 3. LOGICAL OPERATORS ---")
x, y = True, False
print("x =", x, ", y =", y)
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)

print("\n--- 4. ASSIGNMENT OPERATORS ---")
c = 5
print("Start c =", c)
c += 2   # c = c + 2
print("c += 2:", c)
c -= 1   # c = c - 1
print("c -= 1:", c)
c *= 3   # c = c * 3
print("c *= 3:", c)
c /= 2   # c = c / 2
print("c /= 2:", c)
c %= 4
print("c %= 4:", c)

print("\n--- 5. BITWISE OPERATORS (work on bits) ---")
m, n = 6, 3   # 6 = 110, 3 = 011
print("m =", m, "binary:", bin(m))
print("n =", n, "binary:", bin(n))
print("m & n (AND):", m & n)
print("m | n (OR):", m | n)
print("m ^ n (XOR):", m ^ n)
print("~m (NOT):", ~m) # ~x = -(x + 1) bit trickier because Python uses complement's of two
print("m << 1 (Left Shift):", m << 1) # x << k   =   x  2^k
print("m >> 1 (Right Shift):", m >> 1) # x >> k   =   floor(x / 2^k)

print("\n--- 6. MEMBERSHIP OPERATORS ---")
my_list = [1, 2, 3, 4]
print("List:", my_list)
print("3 in list:", 3 in my_list)
print("5 not in list:", 5 not in my_list)

print("\n--- 7. IDENTITY OPERATORS ---")
p = [1, 2, 3]
q = p
r = [1, 2, 3]
print("p is q:", p is q)         # same object
print("p is r:", p is r)         # same value, different object
print("p == r:", p == r)         # values compared
````

## List vs Tuple vs Set vs Dictionary

built-in data structures

````python
# 1. All are non-homogeneous, with different ways of storing the elements. (Note: Sets can't be nested.)

my_list = [1, "apple", 3.14, True]
my_tuple = (2, "banana", False, 7.5)
my_set = {3, "orange", 9.81, (1, 2)}

# bad_set = {1, 2, {3, 4}}  # ERROR: sets cannot contain other sets

print("List:", my_list)
print("Tuple:", my_tuple)
print("Set:", my_set)
````

````python
# 2. Sets & Keys of Dictionaries doesn't allow duplicates, others does allow.

lst = [1, 1, 2, 2]
tup = (3, 3, 4, 4)

st = {1, 1, 2, 2}

d = {"a": 1, "a": 99, "b": 2} # Dict keys remove duplicates (latest assignment kept)

print("List:", lst)
print("Tuple:", tup)
print("Set:", st)
print("Dict:", d)
````

````python
# 3. Tuples are immutable, others are mutable.

my_list = [10, 20]
my_set = {10, 20}
my_dict = {"x": 1}
my_tuple = (10, 20)

# Mutating allowed
my_list[0] = 99
my_set.add(30)
my_dict["y"] = 2

# my_tuple[0] = 99  # ERROR: Attempt to mutate a tuple

print("List:", my_list)
print("Set:", my_set)
print("Dict:", my_dict)
print("Tuple:", my_tuple)
````

````python
# 4. Sets are ordered, others aren't. (Python 3.7 and above => Dictionaries are ordered)

# Set: ordering is arbitrary
st = {"apple", "banana", "cherry"}
print("Set (unordered):", st)

# Dict preserves insertion order (Python 3.7+)
d = {}
d["first"] = 1
d["second"] = 2
d["third"] = 3
print("Dict (ordered):", d)
````
