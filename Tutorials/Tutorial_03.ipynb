1. Calculate the sum of numbers from 1 to n (default n=100) using a while loop.

````python
def sum_upto_n(n):
  sum = 0
  i = 0
  while i<=n:
    sum += i
    i+=1
  return sum

sum_upto_n(100)
````

2. Count the number of vowels in a string using a for loop

````python
vowels = "AEIOUaeiou"
s = "Hello"
count = 0

for character in s:
  if character in vowels:
    count += 1

print(count)
````

````python
def count_vowel(s):
  count = 0
  for c in s:
    if c in "AEIOUaeiou":
      count+=1
  return count

count_vowel("Hello")
````

````python
s = input("Enter the string: ")

count_vowel(s)
````

4. Print numbers from 1 to n, except those divisble by 3 using a for loop and continue statement.

````python
n = 10

for i in range(1,n+1):
  if i%3==0:
    continue
  print(i, end=", ")
````
